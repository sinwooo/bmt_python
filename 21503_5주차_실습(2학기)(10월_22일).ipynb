```python
import csv 

f=open('age.csv','r',encoding='cp949')

data=csv.reader(f)

header=next(data)

result=[]

loc_name=input('알고 싶은 지역 입력 : ')

for row in data:
  
  if loc_name in row[0]:

    for i in row[3:]:

      result.append(int(i))

f.close()

import matplotlib.pyplot as plt

plt.style.use('ggplot')

plt.title('population region plot')

plt.plot(result)

plt.show
```
